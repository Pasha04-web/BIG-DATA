```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
from sklearn.metrics import confusion_matrix
from nltk.tokenize import word_tokenize
from keras.preprocessing import sequence
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.models import Model
from keras.layers import Input, Dense, LSTM, Dropout, concatenate, Embedding
from keras.utils import plot_model
```

```python
# Membaca data
df = pd.read_csv('/content/all_agree.csv')

# Menampilkan 5 data teratas
print(df.head())

# Menghitung jumlah sampel masing-masing class
print(df['label_score'].value_counts())
```

```python
def find_punc(x, punc):
    count = 0
    for char in x:
        if char == punc:
            count += 1
    return count

def find_uppercase(x):
    count = 0
    for char in x:
        if char.isupper():
            count += 1
    return count

def find_lowercase(x):
    count = 0
    for char in x:
        if char.islower():
            count += 1
    return count

def find_number(x):
    return len(re.findall('[0-9]', x))
```

```python
# Contoh penerapan untuk tanda baca tertentu
df['comma_count'] = df['title'].apply(lambda x: find_punc(x, ','))
df['fullstop_count'] = df['title'].apply(lambda x: find_punc(x, '.'))
# (Dan seterusnya untuk tanda baca lainnya sesuai dokumen)

# Menerapkan fitur huruf dan angka
df['uppercase_count'] = df['title'].apply(find_uppercase)
df['lowercase_count'] = df['title'].apply(find_lowercase)
df['number_count'] = df['title'].apply(find_number)
```

```python
import nltk
nltk.download('punkt_tab')
def clean(text):
    # Lowering letters
    text = text.lower()
    # Removing punctuations
    text = re.sub('[..!?: \']', '', text)
    # Tokenization
    word_tokens = word_tokenize(text)
    # Joining words
    text = (' '.join(word_tokens))
    return text

# Menerapkan pembersihan pada kolom title
df['cleaned_title'] = df['title'].apply(clean)
```

```python
# Mendefinisikan fitur (X) dan target (y)
X = df.iloc[:, 3:] # Kolom comma_count sampai cleaned_title
y = df.iloc[:, 2]  # Kolom label_score

# Melakukan splitting dengan rasio 80:20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=77)
```

```python
# Memisahkan fitur teks dan fitur tanda baca
X_train_title = X_train['cleaned_title']
X_test_title = X_test['cleaned_title']
X_train_punc = X_train.drop('cleaned_title', axis=1)
X_test_punc = X_test.drop('cleaned_title', axis=1)

# Integer-encoding menggunakan Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train_title)
X_train_title = tokenizer.texts_to_sequences(X_train_title)
X_test_title = tokenizer.texts_to_sequences(X_test_title)

# Mendefinisikan ukuran kosakata (vocab size)
vocab_size = len(tokenizer.word_index)

# Padding kalimat agar panjangnya seragam (19 kata)
X_train_title = sequence.pad_sequences(X_train_title, maxlen=19)
X_test_title = sequence.pad_sequences(X_test_title, maxlen=19)
```

```python
class_weights = class_weight.compute_class_weight(class_weight='balanced', classes=np.array([0, 1]), y=y_train.values)
class_weight_dict = dict(enumerate(class_weights))
```

```python
tf.random.set_seed(99)

# Membangun layer input dan embedding untuk LSTM
lstm_input = Input(shape=(19,))
embedding_layer = Embedding(input_dim=vocab_size + 1, output_dim=32, input_length=19)(lstm_input)

# (Sesuai Gambar 4.29 di dokumen)
lstm_layer = LSTM(64)(embedding_layer)
dropout_layer = Dropout(rate=0.5)(lstm_layer)
output_dense = Dense(1, activation='sigmoid')(dropout_layer)
lstm_model = Model(inputs=lstm_input, outputs=output_dense)
```

```python
# --- Membangun Input MLP (Input 2 untuk fitur angka/tanda baca) ---
dense_input = Input(shape=(5,)) # Input berisi 5 fitur angka
dense = Dropout(rate=0.5)(dense_input)
dense = Dense(3, activation='sigmoid')(dense)
dense_model = Model(inputs=dense_input, outputs=dense)

# --- Menggabungkan Output LSTM dan MLP ---
# Mengambil output dari model lstm dan model dense
concat = concatenate([lstm_model.output, dense_model.output])
concat = Dense(6, activation='sigmoid')(concat)
concat = Dense(1, activation='sigmoid')(concat) # Layer output akhir [cite: 279]

# Mendefinisikan model akhir dengan 2 input
model = Model(inputs=[lstm_model.input, dense_model.input], outputs=concat)
```

```python
# Kompilasi Model
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['acc'])

# Melatih Model (Training)
# Menggunakan X_train_title untuk LSTM dan X_train_punc untuk MLP [cite: 395]
history = model.fit([X_train_title, X_train_punc], y_train,
                    epochs=4,
                    batch_size=64,
                    class_weight=class_weight_dict, # Menggunakan pembobotan class [cite: 387]
                    validation_data=([X_test_title, X_test_punc], y_test))
```

```python
# Menampilkan grafik akurasi
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Menampilkan grafik loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# Menampilkan ringkasan model di console
model.summary()

# Menyimpan arsitektur model ke dalam file gambar
plot_model(model, to_file='model_architecture.png', show_shapes=True, show_layer_names=True)
```

```python
# Melakukan prediksi pada data test
y_pred = model.predict([X_test_title, X_test_punc])
y_pred = (y_pred > 0.5).astype(int) # Mengubah probabilitas menjadi class 0 atau 1

# Membuat Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix Model')
plt.show()
```
