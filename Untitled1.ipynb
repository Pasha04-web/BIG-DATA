```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
import warnings
warnings.filterwarnings('ignore')

print("Isi folder saat ini:")
print(os.listdir('.'))

if 'Font Dataset Large' in os.listdir('.'):
    print("\n Folder 'Font Dataset Large' DITEMUKAN!")
    print("Jumlah font:", len(os.listdir('Font Dataset Large')))
else:
    print("\n Folder 'Font Dataset Large' BELUM DITEMUKAN")
    print("Pastikan proses download dan unzip sudah selesai.")
```

```python
def take_80_pixels(image):
    """Fungsi sesuai modul untuk mengambil/crop gambar menjadi 80x80 pixel"""
    if image is None:
        return None

    h, w = image.shape[:2]

    if h >= 80 and w >= 80:
        start_h = (h - 80) // 2
        start_w = (w - 80) // 2
        return image[start_h:start_h+80, start_w:start_w+80]
    else:
        return cv2.resize(image, (80, 80))

print("Fungsi take_80_pixels sudah dibuat")
```

```python
base_dir = 'Font Dataset Large'

target_fonts = ['arial', 'brandish', 'consolas', 'didot']

font_list = [f for f in sorted(os.listdir(base_dir)) if f.lower() in target_fonts]


print(f"Total kelas font yang digunakan: {len(font_list)}")
print(f"Daftar font: {font_list}")

X = []
y = []

max_per_class = 1000

print("Sedang memproses gambar... Harap tunggu")

for font in font_list:
    font_path = os.path.join(base_dir, font)
    image_files = os.listdir(font_path)[:max_per_class]

    for img_name in image_files:
        img_path = os.path.join(font_path, img_name)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

        if img is not None:
            img_80 = take_80_pixels(img)
            X.append(img_80)
            y.append(font)

X = np.array(X)
y = np.array(y)

print(f"\nTotal gambar yang berhasil di-load: {X.shape[0]}")
print(f"Shape gambar: {X.shape[1:]}")
```

```python
X = X.astype('float32') / 255.0
X = X.reshape(-1, 80, 80, 1)

le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_cat = to_categorical(y_encoded, len(font_list))

X_train, X_test, y_train, y_test = train_test_split(X, y_cat,
                                                    test_size=0.2,
                                                    random_state=42)

print("Data siap digunakan untuk training")
print(f"X_train shape: {X_train.shape}")
```

```python
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(80,80,1)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(256, activation='relu'),
    Dense(len(font_list), activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(X_train, y_train,
                    validation_split=0.2,
                    epochs=5,
                    batch_size=128)
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(f"\n Akurasi Test Set: {test_acc*100:.2f}%")

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Model Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
print("Sedang melakukan prediksi...")
predictions_raw = model.predict(X_test)
predictions = np.argmax(predictions_raw, axis=1)
y_test_labels = np.argmax(y_test, axis=1)

cm = confusion_matrix(y_test_labels, predictions)
label_names = le.classes_
plt.figure(figsize=(8, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=label_names, yticklabels=label_names)
plt.title('Confusion Matrix - Pengenalan Font')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.show()
```

```python
misses = []
for i in range(len(predictions)):
    if y_test_labels[i] != predictions[i]:
        misses.append(i)

print(f"\nIndeks misklasifikasi: {misses}")
```

```python
if len(misses) > 0:
    num_show = min(len(misses), 2) # Tampilkan 2 sesuai contoh tugas
    fig, axes = plt.subplots(ncols=num_show, figsize=(12, 4))

    # Jika hanya 1 salah, axes bukan array, kita paksa jadi array
    if num_show == 1: axes = [axes]

    for i, idx in enumerate(misses[:num_show]):
        actual = label_names[y_test_labels[idx]]
        pred = label_names[predictions[idx]]

        axes[i].imshow(X_test[idx].reshape(80, 80), cmap='gray')
        axes[i].set_title(f"Actual: {actual}\nPredicted: {pred}")
        axes[i].axis('off')
    plt.show()
else:
    print("Tidak ada misklasifikasi untuk ditampilkan.")
```
